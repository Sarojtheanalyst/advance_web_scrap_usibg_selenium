```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
import numpy as np
# https://www.daraz.com.np/catalog/?q=shoes%20for%20man
title=[]
price=[]
sold_status=[]
discount_per=[]
province=[]
for j in range(1,103):
    print(f"{j} pages scrap sucessfully")
    driver.get(f"https://www.daraz.com.np/catalog/?page={j}&q=shoes%20for%20man")
    time.sleep(10)
    for i in range(1,41):
        try:
            title.append(driver.find_element(By.XPATH, f'/html/body/div[4]/div/div[2]/div[1]/div/div[1]/div[2]/div[{i}]/div/div/div[2]/div[2]').text)# title
        except:
            title.append(np.nan)
        try:
            price.append(driver.find_element(By.XPATH, f'/html/body/div[4]/div/div[2]/div[1]/div/div[1]/div[2]/div[{i}]/div/div/div[2]/div[3]/span').text)# price
        except:
            price.append(np.nan)
        try:
            sold_status.append(driver.find_element(By.XPATH, f'/html/body/div[4]/div/div[2]/div[1]/div/div[1]/div[2]/div[{i}]/div/div/div[2]/div[5]/span[1]/span[1]').text)# sold status
        except:
            sold_status.append(np.nan)
        try:
            discount_per.append(driver.find_element(By.XPATH, f'/html/body/div[4]/div/div[2]/div[1]/div/div[1]/div[2]/div[{i}]/div/div/div[2]/div[4]/span').text)# discount percentage
        except:
            discount_per.append(np.nan)
        #/html/body/div[4]/div/div[2]/div[1]/div/div[1]/div[2]/div[39]/div/div/div[2]/div[4]/span
        try:
            province.append(driver.find_element(By.XPATH, f'/html/body/div[4]/div/div[2]/div[1]/div/div[1]/div[2]/div[{i}]/div/div/div[2]/div[5]/span[2]').text)# province
        except:
            province.append(np.nan)

driver.quit()
```

```python
import pandas as pd
df=pd.DataFrame({'title':title,'price':price,'sold_status':sold_status,'discount_per':discount_per,'province':province})
```

```python
df.to_csv("R:\\django sipalaye\\shoes.csv",index=False)
print(" saves sucessfully")
```
