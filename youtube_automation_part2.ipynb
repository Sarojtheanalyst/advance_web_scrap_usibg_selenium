```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time


driver = webdriver.Chrome()
driver.get("https://www.youtube.com/watch?v=Svp2MXZOXpQ")

time.sleep(10)
#driver.execute_script("window.scrollTo(0, 500);")


def scroll_and_get_height():
    last_height = driver.execute_script("return document.documentElement.scrollHeight")

    while True:
        driver.execute_script("window.scrollTo(500, document.documentElement.scrollHeight);")

        
        time.sleep(30)

        # Get the new height of the page
        new_height = driver.execute_script("return document.documentElement.scrollHeight")

      
        if new_height == last_height:
            break

        last_height = new_height

    return new_height


new_height = scroll_and_get_height()


driver.quit()

# Print the new height
print(new_height)
```
