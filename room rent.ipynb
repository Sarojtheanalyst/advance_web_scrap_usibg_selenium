```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
title=[]
location=[]
price=[]
type=[]

driver = webdriver.Chrome()
for j in range(1,18):
    driver.get(f"https://kothabhada.com/latest-properties?page={j}")
    time.sleep(5)
    for i in range(1,21):
        try:
            title.append(driver.find_element(By.XPATH, f'/html/body/section[2]/div/div[4]/div[{i}]/div/div[2]/a/h4').text)
        except:
            title.append("not found")
        try:
            location.append(driver.find_element(By.XPATH, f"/html/body/section[2]/div/div[4]/div[{i}]/div/div[2]/div/div[1]/div/p").text)
        except:
            location.append("unknown location")
        try:
            price.append(driver.find_element(By.XPATH, f"/html/body/section[2]/div/div[4]/div[{i}]/div/div[2]/div/div[3]/div").text)
        except:
            price.append("price not found")
        try:
            type.append(driver.find_element(By.XPATH, f"/html/body/section[2]/div/div[4]/div[{i}]/div/div[2]/div/div[2]/div").text)
        except:
            type.append("not given")
    print(f"scrap the data {j} pages sucrssfully ")
driver.quit()
```
