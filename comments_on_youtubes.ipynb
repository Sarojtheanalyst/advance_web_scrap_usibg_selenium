```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

driver = webdriver.Chrome()

driver.get(f"https://www.youtube.com/watch?v=nY2NwUlM2sc")
time.sleep(10)  

last_height = driver.execute_script("return document.documentElement.scrollHeight")
counter=0
while True:
    driver.execute_script(f"window.scrollTo({last_height}, document.documentElement.scrollHeight);")
    time.sleep(8)  
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    if new_height == last_height:
        break  
    last_height = new_height

comment=[]
views=driver.find_elements(By.XPATH, f'//*[@id="content-text"]/span')
for view in views:
    comment.append(view.text)

likes=[]
like=driver.find_elements(By.XPATH, f'//*[@id="vote-count-middle"]')
for lik in like:
    likes.append(lik.text)

df=pd.DataFrame({'title':comment,'likes':likes})
df.to_csv("R:\\New folder (2)\\songs_comments\\udi_udi.csv",index=False)


driver.quit()
print(" saves files sucessfully")
          
```
