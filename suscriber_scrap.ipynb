```python
from selenium import webdriver
import time

driver = webdriver.Chrome()
driver.get('https://www.youtube.com/@SandeepSeminars/videos')

# Get the current height of the page
last_height = driver.execute_script("return document.body.scrollHeight")

# Scroll from bottom to top
while True:
    # Scroll up by a specific amount
    driver.execute_script("window.scrollTo(0, 0);")
    
    # Wait for a short time to allow the page to load
    time.sleep(10)

    # Get the new height after scrolling
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    # Break the loop if we have reached the top
    if new_height == last_height:
        break

    last_height = new_height

driver.quit()
```
