```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Set up the WebDriver
driver = webdriver.Chrome()

# Open the initial page
driver.get("https://sipalaya.com/all-courses-2/")
time.sleep(5)

for i in range(1, 16):
    try:
        # Locate the course link element
        course_link_xpath = f'/html/body/div[1]/div[2]/div[1]/div/div/div[2]/div/div[{i}]/div[3]/h3/a'
        
        # Wait for the link to be present
        course_link = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, course_link_xpath))
        )

        # Scroll the element into view
        driver.execute_script("arguments[0].scrollIntoView(true);", course_link)

        # Click the link
        course_link.click()

        # Wait for the new page to load
        time.sleep(5)

        # Fetch the image element text
        image_element = driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div/main/div[2]/div[2]/div[1]/div[2]').text
        print(image_element)

        # Navigate back to the courses page
        driver.get("https://sipalaya.com/all-courses-2/")
        time.sleep(5)  # Wait for the page to load again

    except Exception as e:
        print(f"Exception occurred at index {i}: {e}")
        driver.get("https://sipalaya.com/all-courses-2/")
        time.sleep(4)  # Wait for the page to load

# Close the browser
driver.quit()
```

```python
# down;loads sucessfully   - https://www.tiktok.com/@sanjita_adhikary/video/7425058580845399313


from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import requests

# Set up the driver
driver = webdriver.Chrome()  # or webdriver.Firefox(), etc.

# Navigate to the TikTok video page (replace with the actual URL)
video_url = "https://www.tiktok.com/@smriti_bhattarai2/video/7423001389598510343"  # Replace with the actual video URL
driver.get(video_url)

# Wait for the page to load
time.sleep(5)

# Try to find the video element
try:
    video_element = driver.find_element(By.XPATH, "//video")  # Locate the video element
    video_src = video_element.get_attribute('src')  # Get the video source URL

    if video_src:
        # Download the video using requests
        response = requests.get(video_src)

        # Save the video to a file
        with open("tiktok_video.mp4", "wb") as f:
            f.write(response.content)

        print("Video downloaded successfully!")
    else:
        print("Video source URL is empty. Make sure the video element is correctly targeted.")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the browser
    driver.quit()
```
